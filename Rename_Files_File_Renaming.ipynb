```python
#@markdown <h3>Rename file replacing spaces with underscores</h3>
 
txt = input(' Input text here: ')
final_txt = txt.replace(" ", "_")
print(final_txt)
```

```python
 #@markdown <h3>Rename file with lower case letters only.</h3> 
txt = input(' Input text here: ')
txt_lower = txt.lower()
final_txt = txt_lower.replace(" ", "_")
 
print(final_txt)
```
