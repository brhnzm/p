```python
#@markdown <h2><b>←</b> ឵឵Click Here to compare two texts. </h2>
#@markdown <h5><i>(Coded by Muhammad Borhan Uddin Nezami)</i></h5>
text1 = input('Input text1: ')
text2= input('Input text2: ')
 
exact = text1 == text2
match_ignoring_case = text1.lower() == text2.lower()
 
if exact:
    print('✔️✔️Exact match')
elif match_ignoring_case:
        print ('Match but not in case.')
else:
    print('❌ Do not match')
```
